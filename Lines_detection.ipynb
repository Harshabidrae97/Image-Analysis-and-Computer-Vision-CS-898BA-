```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
#read image
img_src = cv2.imread("C://Image_Analysis_CV//Image_Analysis_CV//images//horizontal_line.png",0)
plt.imshow(img_src,cmap='gray')
#img_src = cv2.imread('sample.jpg')

```

```python
#kernal sensitive to horizontal lines
kernel = np.array([[-1.0, -1.0], 
                   [2.0, 2.0],
                   [-1.0, -1.0]])

#kernel = kernel/(np.sum(kernel) if np.sum(kernel)!=0 else 1)

#filter the source image
img_rst = cv2.filter2D(img_src,-1,kernel)

plt.imshow(img_rst,cmap='gray')
#save result image
cv2.imwrite('result.jpg',img_rst)
```
